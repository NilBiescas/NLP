```python
import nltk
from nltk.corpus import stopwords
import string

import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
def open_file(file_path):
    with open(file_path, 'r', encoding='utf8') as file:
        text = file.read()
    return text

def basic_statistics(text: nltk.text, word_concordance: str, 
                     wrd_similar: str, common_contexts: str) -> None:
    print("Concordances of the word: ", word_concordance)
    text.concordance(word_concordance)
    print("\nSimilar words to: ", wrd_similar)
    text.similar(wrd_similar)
    print("\nCommon contexts of: ", common_contexts)
    text.common_contexts(common_contexts)
    print("\nCollocations: ")
    text.collocations()
    
def token_info(tokens) -> tuple:
    return tokens.__len__(), set(tokens).__len__(), round(len(tokens) /  len(set(tokens)), 2)

alice_txt = open_file('alice.txt')
moby_dick_txt = open_file('moby_dick.txt')
adventures_holmes_txt = open_file('Adventures_Holmes.txt')

alice_tokens = nltk.word_tokenize(alice_txt)
moby_dick_tokens = nltk.word_tokenize(moby_dick_txt)
adventures_holmes_tokens = nltk.word_tokenize(adventures_holmes_txt)
```

```python
# Determine the maximum label width
max_label_width = len("Adventures Sherlock INFO:")

# Define a consistent format for all lines
line_format = "{:<{label_width}} Num tokens: {:<6}, Size Vocab: {:<5}; Lexical diversity: {}"

print(line_format.format("Alice INFO:", *token_info(alice_tokens), label_width=max_label_width))
print(line_format.format("Moby Dick INFO:", *token_info(moby_dick_tokens), label_width=max_label_width))
print(line_format.format("Adventures Sherlock INFO:", *token_info(adventures_holmes_tokens), label_width=max_label_width))
```

```python
text =  "Hello world. How are you?"

# Tokenizing into words
tokens = nltk.word_tokenize(text)
print("The tokens are:", tokens)
print("Number of tokens:", len(tokens))
```

```python
utterance = "Jane lent $100 to Peter early this morning."

# Word tokenization using nltk
tokens_tokenize = nltk.word_tokenize(utterance)
print("The tokens are:", tokens_tokenize)
print("Number of tokens:", len(tokens_tokenize))

# Tokenization using split
tokens_split = utterance.split(" ")
print("The tokens are:", tokens_split)
print("Number of tokens:", len(tokens_split))
```

We can see that the number of tokens varies because of how we're splitting them. When we use split(), it only breaks the text into words, keeping punctuation as part of the words. But with the NLTK tokenizer, punctuation marks are treated as separate tokens, so they get split too.

### Exercise 2

```python
#get all the exercise 2 in a function
def exercise2(book: str, word_concordance:str, word_similar:str, common_contexts:str):

    with open(book, 'r', encoding="utf8") as f:
        text = f.read()

    # Computing the concordance of the word "Sherlock" 
    text = nltk.Text(nltk.word_tokenize(text))
    print(f"\n {'-'*8}  Concordances of the word  {'-'*8}\n\t\t", word_concordance, end="\n")
    text.concordance(word_concordance)

    print(f"\n {'-'*8}  Concordances of the word  {'-'*8}\n\t\t", word_similar,  end="\n")
    text.concordance(word_similar)
    
    # Getting the similar words for "extreme"
    print(f"\n {'-'*8}  Similar words to  {'-'*8}\n\t\t", word_similar,  end="\n")
    text.similar(word_similar)

    # Getting the common contexts for "extreme" and "gathering"
    print(f"\n {'-'*8}  Common contexts of  {'-'*8}\n\t", common_contexts,  end="\n")
    text.common_contexts(common_contexts)
    

    # Collocation of the tokens
    print(f"\n {'-'*8}  Collocations  {'-'*8}\n\t\t",  end="\n")
    text.collocations()
```

```python
exercise2('Adventures_Holmes.txt', 'Sherlock', 'extreme', ['extreme', 'gathering'])
```

### Exercise 3

```python
with open('moby_dick.txt', 'r', encoding='utf8') as f:
    moby = f.read()

tokens = nltk.word_tokenize(moby)
text = nltk.Text(tokens)

# Dispersion plot
tokens_2_plot = ["good", "happy", "strong", "bad", "sad", "weak"]
ax = nltk.draw.dispersion_plot(text, tokens_2_plot)
ax.set_yticklabels(tokens_2_plot[::-1])
plt.show()
```

Dispersion plot for the words America, citizens, democracy, freedom, war, peace, equal, and united on the Inaugural Adress Corpus

```python
from nltk.corpus import inaugural

# Loading the "The inagural Addres Corppus" tokens
tokens_inagural = inaugural.words()
inaugural_text = nltk.Text(tokens_inagural)

# Dispersion plot
tokens_2_plot = ["America", "citizens", "democracy", "freedom", "war", "peace", "equal", "united"]
ax = nltk.draw.dispersion_plot(inaugural_text, tokens_2_plot)
ax.set_yticklabels(tokens_2_plot[::-1])
plt.show()
```

```python
# Focusing in some interesting words over the years
cfd = nltk.ConditionalFreqDist(
          (target, fileid[:4])
          for fileid in inaugural.fileids()
          for w in inaugural.words(fileid)
          for target in ['war', 'america']
          if w.lower().startswith(target))


plt.figure(figsize=(15, 6)) 
# Adding a vertical line at the 1918 year 
plt.axvline(x=30.5, color='red', linestyle='--')
cfd.plot()
plt.show()
```

### Exercise 4

```python
# Counting the times that each token appears in the "The inagural Addres Corppus"
tokens_count = Counter(tokens_inagural)

# Getting the 30 tokens that appear most
tokens_and_count_top30 = sorted(tokens_count.items(), key=lambda x: x[1], reverse=True)[:30]
tokens_top30 = [x[0] for x in tokens_and_count_top30]
count_top30 =  [x[1] for x in tokens_and_count_top30]

# Frequency distribution
plt.bar(tokens_top30, count_top30)
plt.title("Frequency distribution")
plt.xticks(rotation=90)
plt.grid(alpha=0.5)
plt.show()

# Cumulative frequency distribution
frequency_dist = nltk.probability.FreqDist(tokens_inagural)
frequency_dist.plot(30, cumulative=True, title="Cumulative frequency distribution", show=False, percents=True)
plt.ylim(0, 45)
plt.show()
```

```python
# Showing the first 50 hapaxes
print(frequency_dist.hapaxes()[:50])
```

#### Removing stop words

We replicate the same plots but excluding the stopwords

```python
# We get the stopwords
stopwords_set = set(stopwords.words('english')) | set(string.punctuation) | set(["'", '"', "`", "''", "``", "--", "-", "’", "“", "”", "‘", "—"])

# Removing the stopwords from "The inagural Addres Corppus" tokens
tokens_inagural_stopwords = [i.lower() for i in tokens_inagural if not i.lower() in stopwords_set]
frequency_dist_stopwords = nltk.probability.FreqDist(tokens_inagural_stopwords)

tokens_top30, count_top30 = map(list, zip(*frequency_dist_stopwords.most_common(30)))
```

```python
# Frequency distribution
plt.bar(tokens_top30, count_top30)
plt.title("Frequency distribution without stopwords")
plt.xticks(rotation=90)
plt.grid(alpha=0.5)
plt.show()
```

```python
frequency_dist_stopwords.plot(30, cumulative=True, title="Cumulative frequency distribution", show=False, percents=True)
plt.ylim(0, 15)
plt.show()
```
